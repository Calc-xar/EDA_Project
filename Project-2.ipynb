```python
#  This drops the values 4,5,6 for clasificacion_final
com_df.drop(com_df[(com_df['CLASIFICACION_FINAL'] == 4) | (com_df['CLASIFICACION_FINAL'] == 5)  | (com_df['CLASIFICACION_FINAL'] == 6)].index, inplace=True)

#  This sets the values of the Positives and Negatives to 1,0
com_df['CLASIFICACION_FINAL'].mask(com_df['CLASIFICACION_FINAL'] == 7, 0, inplace=True)
com_df['CLASIFICACION_FINAL'].mask(com_df['CLASIFICACION_FINAL'] == 2, 1, inplace=True)
com_df['CLASIFICACION_FINAL'].mask(com_df['CLASIFICACION_FINAL'] == 3, 1, inplace=True)
```

```python
#  This shows the total amount of classes in the dataset after removing those that had values 4,5,6 for test cases.. 
tr = ['Negative', 'Positve'] 
plt.bar(tr, com_df['CLASIFICACION_FINAL'].value_counts(), color ='blue',  width = .1, )
plt.xlabel("Classes")
plt.ylabel("Total")
plt.title("Total vs Class")
plt.show()
```

```python
#  This deletes the rows that have the following values in each column since they are have a small percentage within the dataset

com_df.drop(com_df[(com_df['DIABETES'] == 98)  | (com_df['ASMA'] == 98) | (com_df['UCI'] == 99) | (com_df['TABAQUISMO'] == 98) 
                 | (com_df['RESULTADO_ANTIGENO'] == 97) | (com_df['HIPERTENSION'] == 98) | (com_df['OBESIDAD'] == 98) 
                 ].index, inplace=True)

#  This drops off the cols we wont be using for the model.
com_df.drop(['FECHA_ACTUALIZACION', 'ORIGEN','SEXO', 'ID_REGISTRO', 'MUNICIPIO_RES', 'HABLA_LENGUA_INDIG', 'ENTIDAD_RES' , 'TOMA_MUESTRA_LAB',
             'RESULTADO_LAB', 'TOMA_MUESTRA_ANTIGENO', 'PAIS_NACIONALIDAD', 'PAIS_ORIGEN', 'FECHA_INGRESO', 'FECHA_SINTOMAS', 'FECHA_DEF', 
              'INTUBADO' , 'EMBARAZO', 'NACIONALIDAD', 'INDIGENA',  'MIGRANTE', 'EDAD'
           ],inplace=True, axis=1)

```

```python
#  This drops rows that are duplicates after we remove columns
X = com_df.drop_duplicates()
```

```python
# This is the dataset that we will use for the model
X
```

```python
# Sets Y to the results ( Postive or  Nevative)
Y = X['CLASIFICACION_FINAL']
```

```python
# drops the y col from the rest the dataset left. Since we already have it on Y 
X.drop(['CLASIFICACION_FINAL'], inplace=True, axis=1)
```

```python
#  This shows how the dataset looks like for the columns that we will use for the models.
X
```

```python
#  This shows the total amount for each class for the dataset we will use for the model after deleting the columns

plt.bar(tr, Y.value_counts(), color ='blue',  width = .1, )
plt.xlabel("Classes")
plt.ylabel("Total")
plt.title("Total vs Class")
plt.show()


#  Below we can now see that the classes are about the same now compared to when we first saw the dataset
```

```python
#  Below is for creating the model to predict
from sklearn import tree
from sklearn.model_selection import train_test_split

#  This part should be for breaking the data apart.  ===========  creating the model  ===================
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size= 0.25, random_state=42)

# Decision Tree Model
from sklearn.tree import DecisionTreeClassifier
decision_tree_model = DecisionTreeClassifier( max_depth=30, min_samples_split= 30)
# Train model
decision_tree_model.fit(X_train, y_train)

y_P = decision_tree_model.predict(X_test)
# Test model
decision_tree_model.score(X_test, y_test)

# Prints the accuracy 
print(" Accuracy of the decision tree model is: ", format(decision_tree_model.score(X_test, y_test), ".2f"))
```

```python
# Ensemble 
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score

clf = RandomForestClassifier(n_estimators=10, max_depth= 10, min_samples_split=100)
# train 
clft = clf.fit(X_train, y_train)
y_predict = clf.predict(X_test)

scores = cross_val_score(clf, X, Y, cv=5)

# Prints the accuracy 
print(" Accuracy of the random Forest model is: ", format(scores.mean(), ".2f"))
```

```python
#  This is for the heatmap for the decision tree
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay

CM = confusion_matrix(y_test, y_P, labels=decision_tree_model.classes_)
CM1 = ConfusionMatrixDisplay(confusion_matrix=CM, display_labels=decision_tree_model.classes_)
CM1.plot()
```

```python
CM_1 = confusion_matrix(y_test, y_predict )
CM1_1 = ConfusionMatrixDisplay(confusion_matrix=CM)
CM1_1.plot()
```
